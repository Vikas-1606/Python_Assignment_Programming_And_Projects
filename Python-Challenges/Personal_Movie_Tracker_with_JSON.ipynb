 **Challenge:  Personal Movie Tracker with JSON**

**Create a Python CLI tool that lets users maintain their own personal movie database, like a mini IMDb.**

Your program should:
1. Store all movie data in a `movies.json` file.
2. Each movie should have:
   - Title
   - Genre
   - Rating (out of 10)
3. Allow the user to:
   - Add a movie
   - View all movies
   - Search movies by title or genre
   - Exit the app

Bonus:
- Prevent duplicate titles from being added
- Format output in a clean table
- Use JSON for reading/writing structured data


```python
import json
import os

FILE_NAME = "movies.json"

def load_movies():
    if not os.path.exists(FILE_NAME):
        return []

    with open(FILE_NAME, "r") as file:
        try:
            return json.load(file)
        except json.JSONDecodeError:
            return []

def save_movies(movies):
    with open(FILE_NAME, "w") as file:
        json.dump(movies, file, indent=4)

def add_movie():
    movies = load_movies()

    title = input("Enter movie title: ").strip()

    # Prevent duplicate titles
    for movie in movies:
        if movie["title"].lower() == title.lower():
            print("Movie already exists!")
            return

    genre = input("Enter genre: ").strip()

    try:
        rating = float(input("Enter rating (0–10): "))
        if rating < 0 or rating > 10:
            print("Rating must be between 0 and 10.")
            return
    except ValueError:
        print("Invalid rating!")
        return

    new_movie = {
        "title": title,
        "genre": genre,
        "rating": rating
    }

    movies.append(new_movie)
    save_movies(movies)

    print("Movie added successfully!")

def view_movies():
    movies = load_movies()

    if not movies:
        print("No movies found.")
        return

    print("\nYour Movies:")
    print("-" * 50)
    print(f"{'Title':20} {'Genre':15} {'Rating':10}")
    print("-" * 50)

    for movie in movies:
        print(f"{movie['title']:20} {movie['genre']:15} {movie['rating']:10}")

    print("-" * 50)

def search_movies():
    movies = load_movies()

    keyword = input("Enter title or genre to search: ").strip().lower()

    results = []

    for movie in movies:
        if (keyword in movie["title"].lower() or
            keyword in movie["genre"].lower()):
            results.append(movie)

    if not results:
        print("No matching movies found.")
        return

    print("\nSearch Results:")
    print("-" * 50)
    print(f"{'Title':20} {'Genre':15} {'Rating':10}")
    print("-" * 50)

    for movie in results:
        print(f"{movie['title']:20} {movie['genre']:15} {movie['rating']:10}")

    print("-" * 50)

def main():
    while True:
        print("\n=== Movie Tracker ===")
        print("1. Add Movie")
        print("2. View All Movies")
        print("3. Search Movie")
        print("4. Exit")

        choice = input("Enter your choice: ").strip()

        if choice == "1":
            add_movie()
        elif choice == "2":
            view_movies()
        elif choice == "3":
            search_movies()
        elif choice == "4":
            print("Goodbye!")
            break
        else:
            print("Invalid choice. Try again.")

# Run the program
main()
```
