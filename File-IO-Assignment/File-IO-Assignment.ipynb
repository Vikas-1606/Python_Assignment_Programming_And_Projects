**1. Write a Python program to read an entire text file.**

```python
try:
    with open("sample.txt", "r") as file:
        content = file.read()
        print(content)
except FileNotFoundError:
    print("File not found!")
```

**2. Write a Python program to read first n lines of a file.**

```python
n = int(input("Enter number of lines to read: "))

with open("sample.txt", "r") as file:
    for i in range(n):
        line = file.readline()
        if not line:   # Stop if file ends before n lines
            break
        print(line, end="")
```

**3. Write a Python program to append text to a file and display the text.**

```python
# Take input from user
text = input("Enter text to append: ")

# Append text to file
with open("sample.txt", "a") as file:
    file.write(text + "\n")

# Read and display the file content
with open("sample.txt", "r") as file:
    content = file.read()
    print("\nFile Content:\n")
    print(content)
```

**4. Write a Python program to read last n lines of a file.**

```python
n = int(input("Enter number of last lines to read: "))

with open("sample.txt", "r") as file:
    lines = file.readlines()

# Get last n lines
for line in lines[-n:]:
    print(line, end="")
```

**6. Write a Python program to read a file line by line store it into a variable.**

```python
content = ""

with open("sample.txt", "r") as file:
    for line in file:
        content += line   # store each line in variable

# Display the content
print(content)
```

**7. Write a python program to find the longest words.**

```python
with open("sample.txt", "r") as file:
    words = file.read().split()

# Find the maximum length
max_length = max(len(word) for word in words)

# Get all words with maximum length
longest_words = [word for word in words if len(word) == max_length]

print("Longest word(s):", longest_words)
print("Length:", max_length)
```

**8. Write a Python program to count the number of lines in a text file.**

```python
count = 0

with open("sample.txt", "r") as file:
    for line in file:
        count += 1

print("Number of lines:", count)
```

**9. Write a Python program to count the frequency of words in a file.**

```python
from collections import Counter

with open("sample.txt", "r") as file:
    words = file.read().lower().split()

# Count frequency
word_count = Counter(words)

# Display result
for word, count in word_count.items():
    print(word, ":", count)
```

**10. Write a Python program to get the file size of a plain file.**

**11. Write a Python program to write a list to a file.**

```python
my_list = ["Apple", "Banana", "Mango", "Orange"]

with open("basic.txt", "w") as file:
    for item in my_list:
        file.write(item + "\n")

print("List written to file successfully!")
```

**12. Write a Python program to copy the contents of a file to another file .**

```python
with open("sample.txt", "r") as src:
    content = src.read()

with open("basic.txt", "w") as dest:
    dest.write(content)

print("File copied successfully!")
```

**13. Write a Python program to read a random line from a file.**

```python
import random

with open("sample.txt", "r") as file:
    lines = file.readlines()

# Select a random line
random_line = random.choice(lines)

print("Random line:")
print(random_line)
```

**14. Write a Python program to assess if a file is closed or not.**

```python
file = open("sample.txt", "r")

# Check file status
print("Is file closed?", file.closed)

file.close()

# Check again after closing
print("Is file closed?", file.closed)
```

**15.Write a Python program that takes a text file as input and returns the number of words of a given text file.**

```python
file_name = input("Enter file name: ")

with open(file_name, "r") as file:
    text = file.read()
    words = text.split()

print("Number of words:", len(words))
```
