**1. Write a Python function to find the maximum of three numbers.**

```python
def max_of_three(n1,n2,n3):
    if n1 >n2 and n1>n3:
        return n1
    elif n2 >n1 and n2>n3:
        return n2
    else:
        return n3

n1=int(input("Enter First Number : "))
n2=int(input("Enter Second Number : "))
n3=int(input("Enter Third Number"))
result=max_of_three(n1,n2,n3)
print("The Maximum Number Is : ",result)
```

**2. Write a Python function to sum all the numbers in a list.**

	Sample List : (8, 2, 3, 0, 7)

	Expected Output : 20

```python
def sum_list(l):
    total = 0
    for i in l:
        total += i
    return total

l=[8,2,3,0,7]
result=sum_list(l)
print(result)
```

**4. Write a Python function to reverse a string.**

	Sample String : "1234abcd"
	Expected Output : "dcba4321"

```python
def reverse_s(str):
    result=str[::-1]
    return result

str="1234abcd"
final_op=reverse_s(str)
print("The Reverse String Is ",final_op)
```

**6. Write a Python function that accepts a string and counts the number of upper and lower case letters.**


    Sample String : 'The quick Brow Fox'
	Expected Output :
	No. of Upper case characters : 3
	No. of Lower case Characters : 12

```python
def count_u_l(str):
    u_count=0
    l_count=0

    for char in str:
        if char.isupper():
            u_count=u_count+1
        elif char.islower():
            l_count=l_count+1

    return u_count , l_count

u,l = count_u_l('The quick Brow Fox')

print("No. of Upper case characters :",u )
print("No. of Lower case characters :", l)
```

**7. Write a Python function that takes a list and returns a new list with distinct elements from the first list.**

	Sample List : [1,2,3,3,3,3,4,5]
	Unique List : [1, 2, 3, 4, 5]

```python
def dis(num1):
    dis_l=[]
    for num in num1:
        if num not in dis_l:
            dis_l.append(num)
    return dis_l
# Example
numbers = [1, 2, 2, 3, 4, 4, 5]
result = dis(numbers)
print(result)
```

**8. Write a Python function that checks whether a passed string is a palindrome or not.**

	Note: A palindrome is a word, phrase, or sequence that reads the same backward as forward, e.g., madam or nurses run.

```python
def pal(str):
    r_str=""
    for char in str:
        r_str=char+r_str
    if r_str == str:
        return "Given String Is Palindrome"
    else:
        return "Given String Is Not Palindrome"

# Example
str= 'madam'
print(pal(str))
```

```python
def pal(s):
    if s == s[::-1]:
        return "Palindrome"
    else:
        return "Not Palindrome"

print(pal('madam'))
```

```python
str = "green-red-yellow-black-white"
def h_sep(str):
    words = str.split('-')
    words.sort()
    return print("Words in a hyphen-separated sequence after sorting them alphabetically Order : - ",'-'.join(words))

print(h_sep(str))
```

**13. Write a Python program to sort a list of tuples using Lambda.**

	Original list of tuples:
	[('English', 88), ('Science', 90), ('Maths', 97), ('Social sciences', 82)]
	Sorting the List of Tuples:
	[('Social sciences', 82), ('English', 88), ('Science', 90), ('Maths', 97)]

```python
def s_tuple(data):
    return sorted(data, key=lambda x: x[1])
data = [('English', 88), ('Science', 90), ('Maths', 97), ('Social sciences', 82)]

result = s_tuple(data)
print(result)
```

**19. Write a Python program to rearrange positive and negative numbers in a given array using Lambda.**

	Original arrays:
	[-1, 2, -3, 5, 7, 8, 9, -10]
	Rearrange positive and negative numbers of the said array:
	[2, 5, 7, 8, 9, -10, -3, -1]


```python
def rearrange(arr):
    positive = list(filter(lambda x: x >= 0, arr))
    negative = list(filter(lambda x: x < 0, arr))
    return positive + negative

arr = [-1, 2, -3, 5, 7, 8, 9, -10]
result = rearrange(arr)
print(result)
```

**20. Write a Python program to count the even and odd numbers in a given array of integers using Lambda.**

	Original arrays:
	[1, 2, 3, 5, 7, 8, 9, 10]
	Number of even numbers in the above array: 3
	Number of odd numbers in the above array: 5

```python
def count_even_odd(arr):
    even_count = len(list(filter(lambda x: x % 2 == 0, arr)))
    odd_count = len(list(filter(lambda x: x % 2 != 0, arr)))

    return even_count, odd_count

arr = [1, 2, 3, 5, 7, 8, 9, 10]
even, odd = count_even_odd(arr)

print("Number of even numbers:", even)
print("Number of odd numbers:", odd)
```

**22. Write a Python program to check whether a given string contains a capital letter, a lower case letter, a number and a minimum length using lambda.**

```python
def check_string(s):
    has_upper = lambda s: any(x.isupper() for x in s)
    has_lower = lambda s: any(x.islower() for x in s)
    has_digit = lambda s: any(x.isdigit() for x in s)
    min_length = lambda s: len(s) >= 8

    if has_upper(s) and has_lower(s) and has_digit(s) and min_length(s):
        return "Valid String"
    else:
        return "Invalid String"

# Example
print(check_string("Abc12345"))
print(check_string("abc123"))
```
