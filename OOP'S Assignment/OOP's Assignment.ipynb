**4. Write a Python program to create a class that represents a shape. Include methods to calculate its area and perimeter. Implement subclasses for different shapes like circle, triangle, and square.**

```python
class Shape:
    def area(self):
        pass

    def perimeter(self):
        pass

class Circle(Shape):
    def __init__(self, r):
        self.r = r

    def area(self):
        return 3.14 * self.r * self.r

    def perimeter(self):
        return 2 * 3.14 * self.r

class Square(Shape):
    def __init__(self, side):
        self.side = side

    def area(self):
        return self.side * self.side

    def perimeter(self):
        return 4 * self.side

class Triangle(Shape):
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def perimeter(self):
        return self.a + self.b + self.c

    def area(self):
        # Simple formula (Heron's Formula)
        s = (self.a + self.b + self.c) / 2
        return (s * (s - self.a) * (s - self.b) * (s - self.c)) ** 0.5

c = Circle(5)
print("Circle Area:", c.area())
print("Circle Perimeter:", c.perimeter())

s = Square(4)
print("Square Area:", s.area())
print("Square Perimeter:", s.perimeter())

t = Triangle(3, 4, 5)
print("Triangle Area:", t.area())
print("Triangle Perimeter:", t.perimeter())
```

**5. Write a Python program to create a class representing a binary search tree. Include methods for inserting and searching for elements in the binary tree.**

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BST:
    def __init__(self):
        self.root = None

    def insert(self, data):
        if self.root is None:
            self.root = Node(data)
        else:
            self._insert(self.root, data)

    def _insert(self, node, data):
        if data < node.data:
            if node.left is None:
                node.left = Node(data)
            else:
                self._insert(node.left, data)
        else:
            if node.right is None:
                node.right = Node(data)
            else:
                self._insert(node.right, data)

    def search(self, data):
        return self._search(self.root, data)

    def _search(self, node, data):
        if node is None:
            return "Not Found"
        if node.data == data:
            return "Found"
        elif data < node.data:
            return self._search(node.left, data)
        else:
            return self._search(node.right, data)

# Example
tree = BST()

tree.insert(10)
tree.insert(5)
tree.insert(15)
tree.insert(3)
tree.insert(7)

print(tree.search(7))
print(tree.search(20))
```

**7. Write a Python program to create a class representing a linked list data structure. Include methods for displaying linked list data, inserting and deleting nodes.**

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    # Insert at end
    def insert(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return

        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node

    # Delete a node
    def delete(self, key):
        temp = self.head

        # If head node itself holds the key
        if temp and temp.data == key:
            self.head = temp.next
            temp = None
            print(f"{key} deleted")
            return

        prev = None
        while temp and temp.data != key:
            prev = temp
            temp = temp.next

        if temp is None:
            print("Value not found!")
            return

        prev.next = temp.next
        temp = None
        print(f"{key} deleted")

    # Display list
    def display(self):
        temp = self.head
        if temp is None:
            print("Linked List is empty")
            return

        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")

ll = LinkedList()

ll.insert(10)
ll.insert(20)
ll.insert(30)

print("Linked List:")
ll.display()

ll.delete(20)

print("After deletion:")
ll.display()
```

```python
class ShoppingCart:
    def __init__(self):
        self.cart = {}   # dictionary to store items {item: price}

    # Add item
    def add_item(self, item, price):
        self.cart[item] = price
        print(f"{item} added with price ₹{price}")

    # Remove item
    def remove_item(self, item):
        if item in self.cart:
            del self.cart[item]
            print(f"{item} removed from cart")
        else:
            print("Item not found in cart")

    # Display cart
    def display_cart(self):
        if not self.cart:
            print("Cart is empty")
        else:
            print("\nItems in Cart:")
            for item, price in self.cart.items():
                print(f"{item} : ₹{price}")

    # Calculate total price
    def total_price(self):
        total = sum(self.cart.values())
        print(f"\nTotal Price = ₹{total}")

cart = ShoppingCart()

cart.add_item("Shirt", 500)
cart.add_item("Shoes", 1200)
cart.add_item("Watch", 800)

cart.display_cart()

cart.total_price()

cart.remove_item("Shoes")

cart.display_cart()
cart.total_price()
```

**9. Write a Python program to create a class representing a stack data structure. Include methods for pushing, popping and displaying elements.**

```python
class Stack:
    def __init__(self):
        self.stack = []

    # Push element into stack
    def push(self, item):
        self.stack.append(item)
        print(f"{item} pushed into stack")

    # Pop element from stack
    def pop(self):
        if self.is_empty():
            print("Stack is empty! Cannot pop")
        else:
            removed = self.stack.pop()
            print(f"{removed} popped from stack")

    # Display stack elements
    def display(self):
        if self.is_empty():
            print("Stack is empty")
        else:
            print("Stack elements (top -> bottom):")
            for item in reversed(self.stack):
                print(item)

    # Check if stack is empty
    def is_empty(self):
        return len(self.stack) == 0

s = Stack()

s.push(10)
s.push(20)
s.push(30)

s.display()

s.pop()
s.display()
```

**10. Write a Python program to create a class representing a queue data structure. Include methods for enqueueing and dequeueing elements.**

```python
class Queue:
    def __init__(self):
        self.queue = []

    # Enqueue (add element)
    def enqueue(self, item):
        self.queue.append(item)
        print(f"{item} added to queue")

    # Dequeue (remove element)
    def dequeue(self):
        if self.is_empty():
            print("Queue is empty! Cannot dequeue")
        else:
            removed = self.queue.pop(0)
            print(f"{removed} removed from queue")

    # Check if empty
    def is_empty(self):
        return len(self.queue) == 0

    # Display queue
    def display(self):
        if self.is_empty():
            print("Queue is empty")
        else:
            print("Queue elements (front -> rear):", self.queue)

q = Queue()

q.enqueue(10)
q.enqueue(20)
q.enqueue(30)

q.display()

q.dequeue()
q.display()
```

```python
class BankAccount:
    def __init__(self, accountNumber, name, balance):
        self.accountNumber = accountNumber
        self.name = name
        self.balance = balance

    # Deposit method
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"₹{amount} deposited successfully")
        else:
            print("Invalid deposit amount")

    # Withdrawal method
    def withdrawal(self, amount):
        if amount > self.balance:
            print("Insufficient balance!")
        elif amount <= 0:
            print("Invalid withdrawal amount")
        else:
            self.balance -= amount
            print(f"₹{amount} withdrawn successfully")

    # Bank fees (5%)
    def bankFees(self):
        fee = self.balance * 0.05
        self.balance -= fee
        print(f"Bank fees of ₹{fee} applied")

    # Display account details
    def display(self):
        print("\n--- Account Details ---")
        print(f"Account Number : {self.accountNumber}")
        print(f"Account Holder : {self.name}")
        print(f"Balance        : ₹{self.balance}")

acc = BankAccount(12345, "Vikas", 10000)

while True:
    print("\n1. Deposit\n2. Withdraw\n3. Apply Bank Fees\n4. Display\n5. Exit")
    choice = int(input("Enter choice: "))

    if choice == 1:
        amt = float(input("Enter amount: "))
        acc.deposit(amt)

    elif choice == 2:
        amt = float(input("Enter amount: "))
        acc.withdrawal(amt)

    elif choice == 3:
        acc.bankFees()

    elif choice == 4:
        acc.display()

    elif choice == 5:
        print("Thank you!")
        break

    else:
        print("Invalid choice!")
```

**14. Write a program that uses datetime module within a class. Enter manufacturing date and expiry date of the product. The program must display the years, months and days that are left for expiry.**

```python
from datetime import datetime

class Product:
    def __init__(self, mfg_date, exp_date):
        # Convert string input into datetime objects
        self.mfg_date = datetime.strptime(mfg_date, "%d-%m-%Y")
        self.exp_date = datetime.strptime(exp_date, "%d-%m-%Y")

    def time_to_expiry(self):
        today = datetime.now()

        if self.exp_date < today:
            print("Product already expired ")
            return

        # Difference in days
        diff = self.exp_date - today
        days_left = diff.days

        # Convert days into years, months, days (approx)
        years = days_left // 365
        months = (days_left % 365) // 30
        days = (days_left % 365) % 30

        print(f"Time left for expiry: {years} years, {months} months, {days} days")

mfg = input("Enter Manufacturing Date (DD-MM-YYYY): ")
exp = input("Enter Expiry Date (DD-MM-YYYY): ")

p = Product(mfg, exp)
p.time_to_expiry()
```

	Make a method display to display flavour and price of each Scoop in scoop_list and print total price of the bowl by adding all flavour scoops prices.

	Make a method sold in both Scoop class and Bowl class to print no of quantity sold.

```python
class Scoop:
    __sold_count = 0   # class variable to track scoops sold

    def __init__(self, flavor):
        self.flavor = flavor          # public property
        self.__price = 0              # private property

    # Setter for price
    def set_price(self, price):
        self.__price = price

    # Getter for price
    def get_price(self):
        return self.__price

    # Sold method
    @classmethod
    def sold(cls):
        print(f"Total scoops sold: {cls.__sold_count}")

    # Internal method to increase count
    def _increase_sold(self):
        Scoop.__sold_count += 1

class Bowl:
    __bowl_sold = 0   # class variable

    def __init__(self):
        self.__scoop_list = []   # private list

    # Add multiple scoops
    def add_scoops(self, *scoops):
        for scoop in scoops:
            self.__scoop_list.append(scoop)
            scoop._increase_sold()   # track scoop sale

    # Display details
    def display(self):
        total = 0
        print("\n--- Bowl Details ---")

        for scoop in self.__scoop_list:
            print(f"Flavor: {scoop.flavor}, Price: ₹{scoop.get_price()}")
            total += scoop.get_price()

        print(f"Total Price of Bowl: ₹{total}")

    # Sold method
    @classmethod
    def sold(cls):
        print(f"Total bowls sold: {cls.__bowl_sold}")

    # Call when bowl is sold
    def checkout(self):
        Bowl.__bowl_sold += 1

# Create scoops
s1 = Scoop("Vanilla")
s1.set_price(50)

s2 = Scoop("Chocolate")
s2.set_price(60)

s3 = Scoop("Strawberry")
s3.set_price(55)

# Create bowl
b = Bowl()

# Add scoops
b.add_scoops(s1, s2, s3)

# Display bowl
b.display()

# Checkout bowl
b.checkout()

# Show sold counts
Scoop.sold()
Bowl.sold()
```

**17. Create a Bus child class that inherits from the Vehicle class. The default fare charge of any vehicle is seating capacity * 100. If Vehicle is Bus instance, we need to add an extra 10% on full fare as a maintenance charge. So total fare for bus instance will become the final amount = total fare + 10% of the total fare.**

**Note: The bus seating capacity is 50. so the final fare amount should be 5500. You need to override the fare() method of a Vehicle class in Bus class.**

```python
class Vehicle:
    def __init__(self, name, capacity):
        self.name = name
        self.capacity = capacity

    # Default fare
    def fare(self):
        return self.capacity * 100


class Bus(Vehicle):
    def __init__(self, name, capacity=50):
        super().__init__(name, capacity)

    # Override fare method
    def fare(self):
        base_fare = super().fare()
        maintenance = base_fare * 0.10
        total_fare = base_fare + maintenance
        return total_fare

bus = Bus("School Bus")
print("Total Bus Fare:", bus.fare())
```

```python
class Employee:
    def __init__(self, emp_name, emp_id, emp_salary, emp_department):
        self.emp_name = emp_name
        self.emp_id = emp_id
        self.emp_salary = emp_salary
        self.emp_department = emp_department

    # Assign / change department
    def emp_assign_department(self, new_department):
        self.emp_department = new_department
        print(f"{self.emp_name} is now working in {self.emp_department} department")

    # Calculate salary with overtime
    def calculate_emp_salary(self, hours_worked):
        if hours_worked > 50:
            overtime = hours_worked - 50
            overtime_amount = overtime * (self.emp_salary / 50)
            total_salary = self.emp_salary + overtime_amount
        else:
            total_salary = self.emp_salary

        print(f"Salary of {self.emp_name}: ₹{total_salary}")

    # Print employee details
    def print_employee_details(self):
        print("\n--- Employee Details ---")
        print("Name       :", self.emp_name)
        print("ID         :", self.emp_id)
        print("Salary     :", self.emp_salary)
        print("Department :", self.emp_department)

# Creating employees
e1 = Employee("ADAMS", "E7876", 50000, "ACCOUNTING")
e2 = Employee("JONES", "E7499", 45000, "RESEARCH")
e3 = Employee("MARTIN", "E7900", 50000, "SALES")
e4 = Employee("SMITH", "E7698", 55000, "OPERATIONS")

# Print details
e1.print_employee_details()
e2.print_employee_details()

# Change department
e3.emp_assign_department("HR")

# Calculate salary with working hours
e1.calculate_emp_salary(55)
e2.calculate_emp_salary(48)
```

	Print customer orders.
	Note: Use dictionaries and lists to store the data.

```python
class Restaurant:
    def __init__(self):
        self.menu_items = {}          # {item: price}
        self.book_table = []          # list of table numbers
        self.customer_orders = []     # list of orders (dict)

    # Add item to menu
    def add_item_to_menu(self, item, price):
        self.menu_items[item] = price
        print(f"{item} added to menu with price ₹{price}")

    # Book a table
    def book_tables(self, table_number):
        if table_number not in self.book_table:
            self.book_table.append(table_number)
            print(f"Table {table_number} booked successfully")
        else:
            print(f"Table {table_number} is already booked")

    # Take customer order
    def customer_order(self, table_number, items):
        if table_number not in self.book_table:
            print(f"Table {table_number} is not booked yet!")
            return

        order = {
            "table": table_number,
            "items": items
        }
        self.customer_orders.append(order)
        print(f"Order taken for table {table_number}")

    # Print menu
    def print_menu(self):
        print("\n--- Menu ---")
        for item, price in self.menu_items.items():
            print(f"{item} : ₹{price}")

    # Print booked tables
    def print_tables(self):
        print("\n--- Booked Tables ---")
        print(self.book_table)

    # Print customer orders
    def print_orders(self):
        print("\n--- Customer Orders ---")
        for order in self.customer_orders:
            print(f"Table {order['table']} ordered {order['items']}")

r = Restaurant()

# Add menu items
r.add_item_to_menu("Pizza", 250)
r.add_item_to_menu("Burger", 120)
r.add_item_to_menu("Pasta", 200)

# Book tables
r.book_tables(1)
r.book_tables(2)

# Take orders
r.customer_order(1, ["Pizza", "Burger"])
r.customer_order(2, ["Pasta"])

# Print all data
r.print_menu()
r.print_tables()
r.print_orders()
```

**20. Write a Python class BankAccount with attributes like account_number, balance, date_of_opening and customer_name, and methods like deposit, withdraw, and check_balance.**

```python
class BankAccount:
    def __init__(self, account_number, customer_name, balance, date_of_opening):
        self.account_number = account_number
        self.customer_name = customer_name
        self.balance = balance
        self.date_of_opening = date_of_opening

    # Deposit method
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"₹{amount} deposited successfully")
        else:
            print("Invalid deposit amount")

    # Withdraw method
    def withdraw(self, amount):
        if amount <= 0:
            print("Invalid withdrawal amount")
        elif amount > self.balance:
            print("Insufficient balance ")
        else:
            self.balance -= amount
            print(f"₹{amount} withdrawn successfully")

    # Check balance
    def check_balance(self):
        print(f"Current Balance: ₹{self.balance}")

    # Display account details
    def display(self):
        print("\n--- Account Details ---")
        print("Account Number :", self.account_number)
        print("Customer Name  :", self.customer_name)
        print("Balance        :", self.balance)
        print("Opening Date   :", self.date_of_opening)

# Create account
acc = BankAccount(123456, "Vikas", 10000, "10-04-2026")

# Display details
acc.display()

# Perform operations
acc.deposit(2000)
acc.withdraw(3000)

# Check balance
acc.check_balance()
```

**21. Write a Python class Inventory with attributes like item_id, item_name, stock_count, and price, and methods like add_item, update_item, and check_item_details.**

```python
class Inventory:
    def __init__(self):
        # Store items using dictionary: item_id -> details
        self.items = {}

    # Add new item
    def add_item(self, item_id, item_name, stock_count, price):
        if item_id in self.items:
            print("Item already exists!")
        else:
            self.items[item_id] = {
                "name": item_name,
                "stock": stock_count,
                "price": price
            }
            print(f"Item '{item_name}' added successfully")

    # Update existing item
    def update_item(self, item_id, stock_count=None, price=None):
        if item_id not in self.items:
            print("Item not found!")
            return

        if stock_count is not None:
            self.items[item_id]["stock"] = stock_count

        if price is not None:
            self.items[item_id]["price"] = price

        print("Item updated successfully")

    # Check item details
    def check_item_details(self, item_id):
        if item_id not in self.items:
            print("Item not found!")
        else:
            item = self.items[item_id]
            print("\n--- Item Details ---")
            print("ID    :", item_id)
            print("Name  :", item["name"])
            print("Stock :", item["stock"])
            print("Price :", f"₹{item['price']}")

inv = Inventory()

# Add items
inv.add_item(101, "Laptop", 10, 50000)
inv.add_item(102, "Mouse", 50, 500)

# Check details
inv.check_item_details(101)

# Update item
inv.update_item(101, stock_count=8, price=48000)

# Check again
inv.check_item_details(101)
```
